# 8. 배열 생성

# 10. 배열의 크기 변경

- reshape()
- flatten() : 고차원 -> 저차원
- ravel() : 고차원 -> 저차원

```python
arr = np.arange(12)
print(arr)
print(arr.shape)
```

```python
arr1 = arr.reshape(3, 4)
print(arr1)
print(arr1.shape)

print("-------------------------------------------")

arr1 = arr.reshape(3, -1)    # 열의 수를 모를 때 "-1"로 채워두면 알아서 나눠짐
print(arr1)

print("-------------------------------------------")

arr1 = arr.reshape(-1, 4)    # 행의 수를 모를 때 "-1"로 채워두면 알아서 나눠짐
print(arr1)
```

```python
arr2 = arr.reshape(2, 2, 3)  # (면의 개수, 행의 개수, 열의 개수)
print(arr2)

print("-------------------------------------------")

arr2 = arr.reshape(-1, 2, 3)
print(arr2)

print("-------------------------------------------")

arr2 = arr.reshape(2, -1, 3)
print(arr2)
```

```python
print(arr1.reshape(12))  # 원소의 개수를 알아야 함
print(arr1.flatten())    # 원소의 개수 몰라도 됨
print(arr1.ravel())

print("-----------------------------------------")

print(arr2.reshape(12))
print(arr2.flatten())
print(arr2.ravel())
```

## 1) hstack(), vstack()

- vstack() : 수직으로 합치기, 개수 다르면 합치기 불가능
- hstack() : 수평으로 합치기

```python
#### 1차원 배열 ####

a = np.random.randint(0, 10, (4,))  # 범위내 정수를 배열로 출력
print(a)
b = np.ones(4, dtype="i")
print(b)
print("-------------------------------")
print(np.vstack([a, b]))
print("-------------------------------")
print(np.hstack([a, b]))
```

```python
#### 2차원 배열 ####

a = np.random.randint(0, 10, (1, 4))
b = np.random.randint(0, 10, (1, 4))
print(a, b)
print("-------------------------------")
print(np.vstack([a, b]))
print("-------------------------------")
print(np.hstack([a, b]))
```

## 2) concatenate()

- 프로그램 짤 때 용이, 합치는 방향 변수 지정 가능

```python
#### 1차원 ####

a = np.random.randint(0, 10, (4,))  # 범위내 정수를 배열로 출력
print(a)
b = np.ones(4, dtype="i")
print(b)
print("-------------------------------")
print(np.concatenate([a, b]))
```

## 3) stack(), dstack() : 기준이 3차원


- stack() : 차원증가
- dstack() (=np.stack([a1, a2], axis=2)) :  열(깊이)로만 합침

```python
a1 = np.ones((3, 4))
a2 = np.zeros((3, 4))

a3 = np.stack([a1, a2], axis=0)  # 면으로
print(a3)
print(a3.shape)
print("-------------------------------")
a3 = np.stack([a1, a2], axis=1)  # 행으로
print(a3)
print(a3.shape)
print("-------------------------------")
a3 = np.stack([a1, a2], axis=2)  # 열로
print(a3)
print(a3.shape)
```

```python
a1 = np.ones((3, 4))
a2 = np.zeros((3, 4))

a3 = np.dstack([a1, a2])
print(a3)
print(a3.shape)
```

```python
a1 = np.array([1, 2, 3])
a2 = np.array([4, 5, 6])

a3 = np.dstack([a1, a2])
print(a3)
print(a3.shape)
```

```python
a = np.random.randint(0, 10, (2, 2, 3))
b = np.random.randint(0, 10, a.shape) # 같은 크기일 경우 사용 가능
c = np.random.randint(0, 10, a.shape)

d = np.dstack([a, b, c])
print(d)
print(d.shape)
print("-----------------------------")
e = np.stack([a, b, c], axis=2)
print(e)
print(e.shape)
```

# 17. 정수형 데이터 카운팅

- unique()
- bincount()

## 1) unique()

- 중복 제외한 배열 출력
- np.unique(arr, return_counts=True) : 중복 제외한 배열과 각 원소의 개수 추출

```python
arr = np.array(['a', 'b', 'b', 'c', 'a'])
```

```python
print(len(arr))
print(np.unique(arr), len(np.unique(arr)))
```

```python
idx, cnt = np.unique(arr, return_counts=True)
print(idx, cnt)
```

## 2) bincount()

- 발생하지 않은 사건에 대해서도 카운팅
- np.bincount(arr, minlength=n) : n의 범위내에서 나오지 않은 수도 카운트해서 출력

```python
arr = np.array([1, 1, 2, 2, 2, 3])  # 주사위 눈이라 가정(0~5)

print(np.unique(arr, return_counts=True))

print(np.bincount(arr))
print(np.bincount(arr, minlength=6))
```

```python
file = np.loadtxt("data\ml-1m\ratings.dat", delimiter="::", dtype=np.int64)
```

```python
print(file)
print(file.shape)
```

```python
#### 첫행부터 20행까지 출력

print("     {:10}{:10}{:10}{:5}".format("UserID", "MovieID", "Rating", "Timestamp"))
print(file[:20,])
```

```python
#### 21행부터 30행까지 userid, movieid만 출력

print("{}{}".format("UserID", "MovieID"))
print(file[21:31, :2])
print(file[21:31, [0, 1]])
```

```python
#### 전체 평점에 대한 평균

type(file[:, 2])
file[:, 2].mean()
rat_avg = np.mean(file[:, 2])
print("전체 평점에 대한 평균 :", round(rat_avg, 2))
```

```python
#### 사용자별 평점의 평균 출력(10명만)

User_id = np.unique(file[:,0])

means = []

for id in User_id:
    user = file[file[:, 0] == id, :]
    means.append([id, user[:, 2].mean().round(2)])

print("결과 :", means[:10])
```

```python
file[file[:, 0] == 1].mean(axis = 0)[2]

[[id, file[file[:, 0] == id].mean(axis = 0)[2].round(2)] for id in range(1, 11)]
```

```python
# 결과값을 넘파이를 이용하여 파일로 저장하기

np.savetxt("data/ml-1m/ratings_mean.csv", np.array(means), delimiter=",", fmt="%.2f")
```
